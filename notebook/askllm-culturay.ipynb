```python
from datasets import load_dataset
```

```python
def load_scores(filename):
    with open(filename, "r") as f:
        scores = [[float(score), int(index)] for (score, index) in [line.strip().split("\t") for line in f.readlines()]]
    return scores
```

```python
def print_scores(scores, start, end, dataset, text_len):
    width = len(str(len(scores)))
    for i, (score, index) in enumerate(scores[start:end]):
        text = dataset[index]["text"].replace("\n", "\\n")[:text_len]
        print(f"{start+i:{width}d} | {score:.4f} | {text}")
```

```python
dataset_path = "ontocord/CulturaY"
dataset_lang = "ja"
dataset_split = "train"
cache_dir = "/storage7/askllm/hf_cache"

dataset = load_dataset(dataset_path, name=dataset_lang, split=dataset_split, cache_dir=cache_dir)
print(dataset)
```

```python
score_filename = "culturay_scores_sorted.tsv"

scores = load_scores(score_filename)
print(len(scores))
```

```python
# 5個のチャンクに分ける (20%ずつわけてデータを見る)
num_chunks = 5
chunk_size = len(scores) // num_chunks
num_texts = 100  # 表示するテキストの数
text_len = 500   # 表示するテキストの長さ
```

```python
# トップ周辺のデータを表示
chunk_index = 0
start = chunk_size * chunk_index
end = start + num_texts
print_scores(scores, start, end, dataset, text_len)
```

```python
# トップ20%周辺のデータを表示
chunk_index = 1
start = chunk_size * chunk_index
end = start + num_texts
print_scores(scores, start, end, dataset, text_len)
```

```python
# トップ40%周辺のデータを表示
chunk_index = 2
start = chunk_size * chunk_index
end = start + num_texts
print_scores(scores, start, end, dataset, text_len)
```

```python
# トップ60%周辺のデータを表示
chunk_index = 3
start = chunk_size * chunk_index
end = start + num_texts
print_scores(scores, start, end, dataset, text_len)
```

```python
# トップ80%周辺のデータを表示
chunk_index = 4
start = chunk_size * chunk_index
end = start + num_texts
print_scores(scores, start, end, dataset, text_len)
```

```python
# 最下位周辺のデータを表示
end = len(scores)
start = end - num_texts
print_scores(scores, start, end, dataset, text_len)
```

```python
# スコアとデータ数の関係
for i in range(10):
    threshold = 1.0 - (i + 1) * 0.1
    num_above = len([score for score, _ in scores if score > threshold])
    print(f"threshold: {threshold:.1f} | num_above: {num_above:8d} | fraction: {num_above / len(scores):.4f}")
```
